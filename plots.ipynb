```python
import csv
import numpy as np
import pylab as pl
```

```python
# patent

files=[{"type":"patent",
       "metric": "euclidean",
       "files":["./VerifyClusterPatents/section_euclidean_distances.csv",
        "./VerifyClusterPatents/class_euclidean_distances.csv",
        "./VerifyClusterPatents/subclass_euclidean_distances.csv"]},

        {"type":"patent",
       "metric": "cosine",
       "files":["./VerifyClusterPatents/section_cosine_distances.csv",
        "./VerifyClusterPatents/class_cosine_distances.csv",
        "./VerifyClusterPatents/subclass_cosine_distances.csv"]},


        {"type":"paper",
       "metric": "euclidean",
       "files":["./VerifyClusterPapers/Clustering/macroarea_euclidean_distances.csv",
        "./VerifyClusterPapers/Clustering/subject_euclidean_distances.csv",
        "./VerifyClusterPapers/Clustering/topic_euclidean_distances.csv"]},

        {"type":"paper",
       "metric": "cosine",
       "files":["./VerifyClusterPapers/Clustering/macroarea_cosine_distances.csv",
        "./VerifyClusterPapers/Clustering/subject_cosine_distances.csv",
        "./VerifyClusterPapers/Clustering/topic_cosine_distances.csv"]}]
```

```python
pos=[1,2,3,4,5,6,7,8,9]
start=[0,2,5]
end=[2,5,9]

for file in files:
    i=0
    pl.figure(figsize=(20,8))
    for f in file["files"]:
        data=list(csv.reader(open(f,'r'),delimiter=','))
        cut=i+2
        labels=data[0][cut:]
        data=[[np.nan if x=="" else x for x in row] for row in data]
        data=np.array([row[cut:] for row in data[1:]],dtype=float)
        data=data[sum(np.transpose(np.isnan(data)))==0,]
        pl.boxplot(data,labels=labels,positions=pos[start[i]:end[i]])
        pl.title(file["type"]+" "+file["metric"])
        i=i+1
    pl.grid(which='both')
```
